1. Given the [dataset](Datasets/Q1_dataset.csv) which contains the close price for a given stock, can you plot confidence intervals for the data set? The output should be a plot with the stock price and confidence intervals shown on the same plot.

```python
import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
```

```python
stock_price = pd.read_csv('Q1_dataset.csv')
```

```python
#make sure data is fully filled
stock_price.info()
```

```python
stock_price.head()
```

```python
#calculate the standard error using all data points with 95% confidence interval
z_critical = stats.norm.ppf(q = 0.975)
standard_error = z_critical * stock_price['Close'].sem()
#calculate the upper and lower bound
lower_bound = stock_price['Close']-standard_error
upper_bound = stock_price['Close']+standard_error

#Plotting.
plt.plot(stock_price['Close'], linewidth=1) #mean curve.
plt.fill_between(stock_price['Close'].index, lower_bound, upper_bound, color='b', alpha=.2)
plt.xlabel('Date')
plt.ylabel('Price')
```

3. Given this [file](Datasets/Q3_dataset.csv), can you create a cumulative distribution chart of the total years worked across all employees surveyed? You can exclude values that indicate the employee did not state his/her start or end date. Note: Total years worked = Cease Date Year - DETE Start Date Year

```python
#read the dataset
dete_df = pd.read_csv('Datasets/Q3_dataset.csv')
dete_df.head()
```

```python
#check column formats
dete_df['Cease Date'].unique()
```

```python
# We need to clean column formats into years
for index, row in dete_df.iterrows():
    new_row = row['Cease Date'].split('/')
    if len(new_row) == 2:
        dete_df.loc[index,'Cease Date'] = new_row[1]
        
    elif new_row[0].isdigit():
        dete_df.loc[index,'Cease Date'] = new_row[0]
    else:
        continue
```

```python
#check again
dete_df['Cease Date'].unique()
```

```python
#start years are clean
dete_df['DETE Start Date'].unique()
```

```python
#calculate the work years
df_years = dete_df[(~dete_df['Cease Date'].str.contains("Not Stated")) & (~dete_df['DETE Start Date'].str.contains("Not Stated"))][['Cease Date','DETE Start Date']]
df_years = df_years.astype('int')
df_years['Work Years'] = df_years['Cease Date'] - df_years['DETE Start Date']
```

```python
#histogram
df_years['Work Years'].plot.hist(density=True, bins=50)
```

```python
#cumulative density
df_years['Work Years'].hist(cumulative=True, density=True, bins=50)
```
